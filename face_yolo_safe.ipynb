```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/drive/MyDrive/yolo_face_project/face.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    optimizer="AdamW",
    lr0=0.001,
    patience=10,
    device=0,
    project="/content/drive/MyDrive/yolo_face_project/runs",
    name="face_detect"
)
```

```python
import os, shutil, random

os.makedirs(f"{BASE}/images/train", exist_ok=True)
os.makedirs(f"{BASE}/images/val", exist_ok=True)
os.makedirs(f"{BASE}/labels/train", exist_ok=True)
os.makedirs(f"{BASE}/labels/val", exist_ok=True)

images = [f for f in os.listdir(f"{BASE}/images") if f.endswith(".jpg")]
random.shuffle(images)

split = int(0.8 * len(images))
train_imgs, val_imgs = images[:split], images[split:]

for img in train_imgs:
    shutil.move(f"{BASE}/images/{img}", f"{BASE}/images/train/{img}")
    shutil.move(f"{BASE}/labels/{img.replace('.jpg','.txt')}",
                f"{BASE}/labels/train/{img.replace('.jpg','.txt')}")

for img in val_imgs:
    shutil.move(f"{BASE}/images/{img}", f"{BASE}/images/val/{img}")
    shutil.move(f"{BASE}/labels/{img.replace('.jpg','.txt')}",
                f"{BASE}/labels/val/{img.replace('.jpg','.txt')}")

print("✅ Correct train/val split done")
```
